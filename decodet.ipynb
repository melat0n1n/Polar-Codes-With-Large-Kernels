```python
import numpy as np
```

```python
def u_mes_all_1(first_bit):
    val_arr = [0, 1]
    u_mes_array = []
    for i in val_arr:
        for j in val_arr:
            for k in val_arr:
                u_mes_array.append([i,j,k,first_bit])
    return u_mes_array

def u_mes_all_2(first_bit, second_bit):
    val_arr = [0, 1]
    u_mes_array = []
    for i in val_arr:
        for j in val_arr:
            u_mes_array.append([i,j,second_bit,first_bit])
    return u_mes_array

def u_mes_all_3(first_bit, second_bit, third_bit):
    val_arr = [0, 1]
    u_mes_array = []
    for i in val_arr:
            u_mes_array.append([i,third_bit,second_bit, first_bit])
    return u_mes_array
```

```python
def bit_from_llr(value):
    if np.log(value) > 1:
        bit = 1
    else:
        bit = 0
    return bit
```

```python
def sc_decode_kernel(received_mess, gen_matrix):
    decoded_message = []
    
    # find first bit
    u_mes_array_num_1 = u_mes_all_1(0)
    u_mes_array_denom_1 = u_mes_all_1(1)
    x_mes_num = []
    x_mes_denom = []
    for i in range(len(u_mes_array_num_1)):
        x_mes_num.append(np.dot(u_mes_array_num_1[i], gen_matrix))
        x_mes_denom.append(np.dot(u_mes_array_denom_1[i], gen_matrix))
    # find indexes of zero elements
    llr_num = 0
    llr_denom = 0
    for i in range(len(x_mes_num)):
        llr_num_item = 1
        for j in range(len(x_mes_num[i])):
            if x_mes_num[i][j] == 0:
                llr_num_item *= np.exp(received_mess[j])
        llr_num += llr_num_item
    for i in range(len(x_mes_denom)):
        llr_denom_item = 1
        for j in range(len(x_mes_denom[i])):
            if x_mes_denom[i][j] == 0:
                llr_denom_item *= np.exp(received_mess[j])
        llr_denom += llr_denom_item
    decoded_message.append(bit_from_llr(llr_num/llr_denom))
    
    # find second bit
    u_mes_array_num_2 = u_mes_all_2(decoded_message[0], 0)
    u_mes_array_denom_2 = u_mes_all_2(decoded_message[0], 1)
    x_mes_num = []
    x_mes_denom = []
    for i in range(len(u_mes_array_num_2)):
        x_mes_num.append(np.dot(u_mes_array_num_2[i], gen_matrix))
        x_mes_denom.append(np.dot(u_mes_array_denom_2[i], gen_matrix))
    # find indexes of zero elements
    llr_num = 0
    llr_denom = 0
    for i in range(len(x_mes_num)):
        llr_num_item = 1
        for j in range(len(x_mes_num[i])):
            if x_mes_num[i][j] == 0:
                llr_num_item *= np.exp(received_mess[j])
        llr_num += llr_num_item
    for i in range(len(x_mes_denom)):
        llr_denom_item = 1
        for j in range(len(x_mes_denom[i])):
            if x_mes_denom[i][j] == 0:
                llr_denom_item *= np.exp(received_mess[j])
        llr_denom += llr_denom_item
    decoded_message.append(bit_from_llr(llr_num/llr_denom))
    
    # find third bit
    u_mes_array_num_3 = u_mes_all_3(decoded_message[0], decoded_message[1], 0)
    u_mes_array_denom_3 = u_mes_all_3(decoded_message[0], decoded_message[1], 1)
    x_mes_num = []
    x_mes_denom = []
    for i in range(len(u_mes_array_num_3)):
        x_mes_num.append(np.dot(u_mes_array_num_3[i], gen_matrix))
        x_mes_denom.append(np.dot(u_mes_array_denom_3[i], gen_matrix))
    # find indexes of zero elements
    llr_num = 0
    llr_denom = 0
    for i in range(len(x_mes_num)):
        llr_num_item = 1
        for j in range(len(x_mes_num[i])):
            if x_mes_num[i][j] == 0:
                llr_num_item *= np.exp(received_mess[j])
        llr_num += llr_num_item
    for i in range(len(x_mes_denom)):
        llr_denom_item = 1
        for j in range(len(x_mes_denom[i])):
            if x_mes_denom[i][j] == 0:
                llr_denom_item *= np.exp(received_mess[j])
        llr_denom += llr_denom_item
    decoded_message.append(bit_from_llr(llr_num/llr_denom))
    
    # find fourth bit
    u_mes_array_num_4 = np.copy(decoded_message)
    u_mes_array_num_4 = np.append(u_mes_array_num_4, 0)
    u_mes_array_denom_4 = np.copy(decoded_message)
    u_mes_array_denom_4 = np.append(u_mes_array_denom_4, 1)
    x_mes_num = np.dot(u_mes_array_num_4, gen_matrix)
    x_mes_denom = np.dot(u_mes_array_denom_4, gen_matrix)
    # find indexes of zero elements
    llr_num = 0
    llr_denom = 0
    for i in range(len(x_mes_num)):
        if x_mes_num[i] == 0:
            llr_num += np.exp(received_mess[i])
        if x_mes_denom[i] == 0:
            llr_denom += np.exp(received_mess[i])
            
    decoded_message.append(bit_from_llr(llr_num/llr_denom))
    return decoded_message
```
